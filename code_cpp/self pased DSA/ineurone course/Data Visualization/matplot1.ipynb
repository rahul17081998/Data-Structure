```python
# Import dependencies

import numpy as np
import pandas as pd
# Import Matplotlib

import matplotlib.pyplot as plt 
```

```python
x1 = np.linspace(0, 10, 100)
# create a plot figure
fig = plt.figure()

plt.plot(x1, np.sin(x1), '_')
plt.plot(x1, np.cos(x1), '--')
plt.show() # to show the plot
```

```python
plt.figure(figsize=(7,7))
x10 = [35, 25, 20, 10, 23, 234,3,43,3,5]
# labels = ['Computer', 'Electronics', 'Mechanical', 'Chemical']
plt.pie(x10)
plt.show()
fig.savefig('plot1.png')
```

```python
plt.bar(range(len(x10)), x10)
plt.show()
```
